## Introduction

### Méthodes spéciales

Les méthodes spéciales, aussi appelées méthodes magiques, sont des méthodes prédéfinies du langage. Elles permettent de définir le comportement des objets lorsque vous interagissez avec eux par l'intermédiaire d'opérations standards telles que l'addition, l'affichage ou encore la comparaison. Vous avez déjà rencontré des méthodes `comme __init__()` (constructeur de la classe) et `__str__()` (représentation de l'objet sous forme de chaîne de caractères) dans les séances précédentes. Voici d'autres méthodes spéciales :

- `__repr__(self)` : représentation officielle de l'objet, utilisée pour le débogage.
- `__eq__(self, other)` : pour définir l'égalité entre deux objets (`==`).
- `__add__(self, other)` : pour définir l'addition de deux objets (`+`).
- `__sub__(self, other)` : pour définir la soustraction de deux objets (`-`).

#### Exemple

<div style="
  padding: 5pt;
  border-style: dashed;
  border-width: 1px;
  border-color: gray;">

```python
class Vecteur2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        if isinstance(other, Vecteur2D):
            return Vecteur2D(self.x + other.x, self.y + other.y)
        raise ValueError("L'addition nécessite deux vecteurs 2D.")

    def __str__(self):
        return f"({self.x}, {self.y})"
```

</div>

Dans cet exemple, nous définissons le comportement de l'addition entre deux objets de type `Vecteur2D` ainsi que leur affichage sous la forme `(x, y)`. À présent, définissons deux objets du type `Vecteur2D` : `v1` et `v2`, affichons-les et calculons leur somme :

<div style="
  padding: 5pt;
  border-style: dashed;
  border-width: 1px;
  border-color: gray;">

```python
v1 = Vecteur2D(5, 2)
v2 = Vecteur2D(2, 10)
print(v1)
print(v2)
somme = v1 + v2
print(somme)
```

</div>

L'affichage en sortie sera le suivant :

```default
(5, 2)
(2, 10)
(7, 12)
```

### Gestion des exceptions personnalisées

Dans certaines situations, il est utile de créer vos propres exceptions afin de mieux gérer les erreurs spécifiques à votre domaine. Les exceptions personnalisées en Python sont créées en héritant de la classe `Exception`.

#### Exemple

<div style="
  padding: 5pt;
  border-style: dashed;
  border-width: 1px;
  border-color: gray;">

```python
class ErreurCapteur(Exception):
    pass

class ErreurTemperatureInvalide(ErreurCapteur):
    def __init__(self, message="Température invalide détectée"):
        super().__init__(message)
```

</div>

Cette approche permet de lever des erreurs précises lorsque des valeurs erronées sont rencontrées, par exemple une température en dehors des limites admissibles d'un capteur.

### Décorateurs

Un décorateur est une fonction qui modifie ou enrichit le comportement d'une autre fonction ou méthode. Les décorateurs de classes permettent d'ajouter des fonctionnalités à une fonction sans modifier directement son code.

**Syntaxe** :

```python
@nom_decorateur
def ma_fonction():
    pass
```

Exemple de décorateur pour mesurer le temps d'exécution d'une méthode :

<div style="
  padding: 5pt;
  border-style: dashed;
  border-width: 1px;
  border-color: gray;">

```python
import time

def mesure_temps(func):
    def wrapper(*args, **kwargs):
        debut = time.time()
        result = func(*args, **kwargs)
        fin = time.time()
        print(f"Temps d'exécution de {func.__name__}: {fin - debut:.4f} secondes")
        return result
    return wrapper

# Décorateur appliqué à une méthode
class Capteur:
    @mesure_temps
    def lire_valeur(self):
        time.sleep(0.5)  # Simule une lecture
        return 42
```

</div>

Dans cet exemple, le décorateur `mesure_temps` mesure le temps d'exécution de la méthode `lire_valeur()`.

### Classes abstraites et méthodes abstraites

Comme nous l'avons vu dans la séance précédente, les classes abstraites sont des classes qui ne sont pas censées être instanciées directement. Elles définissent une interface que les classes enfants doivent implémenter. En Python, on utilise le module `abc` pour créer des classes et méthodes abstraites.

#### Exemple

<div style="
  padding: 5pt;
  border-style: dashed;
  border-width: 1px;
  border-color: gray;">

```python
from abc import ABC, abstractmethod

class Composant(ABC):
    @abstractmethod
    def calculer_valeur(self):
        pass

class Resistance(Composant):
    def __init__(self, valeur_ohms):
        self.valeur = valeur_ohms

    def calculer_valeur(self):
        return self.valeur
```

</div>

Dans cet exemple, la méthode `lire_valeur()` est déclarée comme abstraite dans la classe `Capteur` et doit être implémentée dans les classes enfants comme `CapteurTemperature`.
